```python
import numpy as np
import pandas as pd
from goldenSplit import goldenSplit
from searchInterval import searchInterval
```

```python
def fun(x, rk):
    return -x[0] + x[1] + rk*(1/(x[1] - 1)**2) + ((x[0] + x[1] - 1)**2)/(rk**0.5)
# 梯度
def objGradient(x):
    g = np.array([-1 + 2*(x[0] + x[1] - 1)/(rk**0.5), 1-2*rk/(x[1]-1)**3 + 2*(x[0] + x[1] -1)/(rk**0.5)])
    gNorm = np.linalg.norm(g, axis = 0)
    if isinstance(gNorm, np.ndarray):
        print(g, '\n', gNorm)
    return g, gNorm

# 迭代主体
def gradientDescent(x0, epsilon, m):
    x = x0
    X, G, Norm = [], [], []
    g, NormG = objGradient(x)
    X.append(x)
    G.append(g)
    ccc = 1
    while NormG >= epsilon and ccc <= m:
        t = searchInterval(x=x, rk=rk, direction=g, func=fun, t0 = 0, h0 = 0.2, alpha = 1.1)
        print('t', t)
        tk = goldenSplit(x, t, epsilon=0.001, direction=g, rk=rk, func = fun)
        print('tk', tk, 'g', g)
        print('x', x)
        print('f', fun(x, rk))
        x = x - tk*g
        X.append(x)
        g, NormG = objGradient(x)
        G.append(g)
        ccc += 1
    return X[-1], G
```

```python
c = 0.1
rk = 10
X = np.array([-1, 2])
epsilon = 1e-5
m = 7
cc = 1
while rk/(X[1] - 1)**2 + (X[0] + X[1] - 1)**2/(rk**0.5) > epsilon and cc <= m:
    print('第%d轮' % (cc))
    Xk, G = gradientDescent(X, 0.00001, m = 5)# 指定内循环最多迭代次数为20
    if np.linalg.norm(Xk-X) < epsilon:
        break
    X = Xk
    rk*=c
    cc+=1
```
